```python
# Modelo de seales
import numpy as np
import matplotlib.pyplot as plt
# INGRESO
a  = -5 # rango de tiempo
b  = 5
dt = 0.1
sigma = 0  # s = 0 +0j
omega = 0
# PROCEDIMIENTO
t = np.arange(a, b, dt)
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(1)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un exponencial simple sigma>0 omega=0
sigma = 1
omega = 0
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(2)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un exponencial simple sigma<0 omega=0
sigma = -1
omega = 0
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(3)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un sinusiode sigma=0, omega>0
sigma = 0
omega = 1
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(4)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un sinusiode sigma=0, omega<0
sigma = 0
omega = -1
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(5)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un sinusoide con amplitud exponencial sigma=0.25, omega>0
sigma = 0.25
omega = 4
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(6)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
# un sinusoide con amplitud exponencial sigma=-0.25, omega>0
sigma = -0.25
omega = 4
s = complex(sigma,omega)
senal = np.exp(s*t)
# SALIDA - grfica
plt.figure(7)
plt.plot(t,np.real(senal),label='real')
plt.plot(t,np.imag(senal),label='imaginaria')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.margins(0.1)
plt.title('e^('+str(sigma) +' + '+str(omega)+'j)')
plt.show()
```
